```python
import numpy as np
import pandas as pd
import os
import glob

# %matplotlib inline
```

```python
#this cell is tagged as parameters
file = "./input_dataset.zip"
outputFilePath = './output_files'
```

```python
def preprocess(s):
    mapping = {k: v for v, k in enumerate(s.batterystate_deviceinfo.unique())}
    s['bstate_deviceinfo'] = s.batterystate_deviceinfo.map(mapping)
    s.time = pd.to_datetime(s.time)
    s.time = s.time.dt.round('1min')
    s = s.drop_duplicates('time',keep='last')
    s.index = s.time
    s = s.drop(s.columns[[0, 1, 2, 3]], axis=1)
    #Last 6 rows of d6 are after 24 hrs....that is not good for LSTM..so dropping it
    s.drop(s.tail(6).index,inplace=True)
    df = s.resample('15min').interpolate(method='linear')
    df = df.drop(df.index[0])
    df.head()
    return df
```

```python
!rm -rf ./input_files; \
rm -rf ./output_files; \
mkdir -p ./input_files; \
mkdir -p ./output_files; \
unzip {file} -d ./input_files;
```

```python
import pandas as pd
from fbprophet import Prophet
import pickle

filename = "./input_files/d4.csv"
s = pd.read_csv(filename, index_col=None, header=0)
df = preprocess(s)
df_pr = df.copy()
df_pr = df.reset_index()
df_pr = df_pr.drop(df_pr.columns[[1, 3, 4]], axis=1)
df_pr.columns = ['ds','y'] # To use prophet column names should be like that
m = Prophet()
m.fit(df_pr)

future = m.make_future_dataframe(periods=2)
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
fig1 = m.plot(forecast)
```

```python
all_files = glob.glob("./input_files/*.csv")

df_list = []
model_list = []

for filename in all_files:
    s = pd.read_csv(filename, index_col=None, header=0)
    df = preprocess(s)
    df_pr = df.copy()
    df_pr = df.reset_index()
    df_pr = df_pr.drop(df_pr.columns[[1, 3, 4]], axis=1)
    df_pr.columns = ['ds','y'] # To use prophet column names should be like that
    m = Prophet()
    m.fit(df_pr)
    m.history = 0
    #m.history_dates = 0
    m.stan_backend = None
    df_list.append(df_pr)
    model_list.append(m)

with open(outputFilePath + '/forecast_model_list.pckl', 'wb') as fout:
    pickle.dump(model_list, fout)
    
```
